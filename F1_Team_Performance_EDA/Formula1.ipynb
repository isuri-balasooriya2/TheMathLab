```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

conn = mysql.connector.connect(
    host="localhost",      
    user="root",       
    password="123456", 
    database="f1"  # 
)


top_constructors = pd.read_sql("SELECT ConstructorName, COUNT(*) as Wins FROM f1results WHERE Position=1 GROUP BY ConstructorName ORDER BY Wins DESC LIMIT 5",conn)

conn.close()

# Plot
plt.figure(figsize=(8,5))
sns.barplot(data=top_constructors,x="ConstructorName", y="Wins", hue ="ConstructorName",palette="Reds_r",legend=False)
plt.title("Top 5 Constructors by Wins")
plt.ylabel("Total Wins")
plt.xlabel("Constructor")
plt.xticks(rotation=45)
plt.show()
```

```python
conn = mysql.connector.connect(
    host="localhost",      
    user="root",       
    password="123456", 
    database="f1"  
)


top_drivers = pd.read_sql("SELECT DriverName, COUNT(*) as Wins FROM f1results WHERE Position=1 GROUP BY DriverName ORDER BY Wins DESC LIMIT 5",conn)

conn.close()

# Plot
plt.figure(figsize=(8,5))
sns.barplot(data=top_drivers,x="DriverName", y="Wins", hue ="DriverName",palette="Blues_r",legend=False)
plt.title("Top 5 Drivers by Wins")
plt.ylabel("Total Wins")
plt.xlabel("Driver")
plt.xticks(rotation=45)
plt.show()

```

```python
conn = mysql.connector.connect(
    host="localhost",      
    user="root",       
    password="123456", 
    database="f1"  
)


ferrari_wins = pd.read_sql("SELECT Season,COUNT(*) as Wins FROM f1results WHERE Position=1 AND ConstructorName = 'Ferrari' GROUP BY Season",conn)

conn.close()

# Plot
plt.figure(figsize=(10, 5))
plt.plot(ferrari_wins["Season"], ferrari_wins["Wins"], marker="o", linestyle="-", color="red")
plt.title("Ferrari Performance Over the Years")
plt.xlabel("Year")
plt.ylabel("Total Wins")
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

```

```python
conn = mysql.connector.connect(
    host="localhost",      
    user="root",       
    password="123456", 
    database="f1"  
)


grid_vs_actual = pd.read_sql("SELECT DriverName, Grid, Position FROM f1results WHERE Season=2024 ORDER BY Grid",conn)

conn.close()

# Plot
plt.figure(figsize=(10, 6))

# Plot Grid Position (starting position)
plt.scatter(grid_vs_actual["Grid"], grid_vs_actual["Grid"], 
            alpha=0.5, color="blue", label="Grid Position", marker="o")

# Plot Actual Position (finishing position)
plt.scatter(grid_vs_actual["Grid"], grid_vs_actual["Position"], 
            alpha=0.5, color="red", label="Actual Position", marker="o")

# Add a title and labels
plt.title("Grid Position vs Actual Position")
plt.xlabel("Grid Position")
plt.ylabel("Position (Actual Finish)")

# Add a legend
plt.legend()

# Add gridlines for better clarity
plt.grid(True)

# Show the plot
plt.show()

```
